```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
def isLatin(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True

```

```python
def cleanTweet(t):
    t=re.sub('#', '', t)
    t=re.sub('[^0-9a-zA-Z]+', ' ', t)
    
    if t.isupper(): t= t.lower() 
    e=0
    for indice, letter in enumerate(t):
        
        if letter.isupper():
            if indice!=0:
                t=t[:indice+e] + ' ' + t[indice+e:]
                e+=1
    g=0
    """
    for indice, letter in enumerate(t):
        
        if is_number(letter):
            if is_number(t[indice-1])==False:
                t=t[:indice+g] + ' ' + t[indice+g:]
                g+=1
                break
            if is_number(t[indice+1])==False:
                t=t[:indice+g] + ' ' + t[indice+g:]
                g+=1
    """
    t=t.lower()
    return t
```

```python
def translateTweet(t):
    t=re.sub('#', '', t)
    traducido=translator.translate(t)
    print(t, traducido.text)
    return traducido.text
```

```python
data2 =pd.read_json(path_json+'2020-03-23_1.json')
df2=pd.DataFrame(data2)
durations=[]
tweets=[]
tweets_volumes=[]
for column in df2:
    if df2.loc['tweet_volume',column]==-1:
        continue

    if isLatin(column):
        tweet_limpio=cleanTweet(column)
        tweet_traducido=translateTweet(tweet_limpio)
        tweets_volumes.append(df2.loc['tweet_volume',column])
        durations.append(df2.loc['duration',column])
        tweets.append(tweet_traducido)
    else:
        tweet_traducido=translateTweet(column)
        tweet_limpio=cleanTweet(tweet_traducido)
        if isLatin(tweet_limpio):
            tweets_volumes.append(df2.loc['tweet_volume',column])
            durations.append(df2.loc['duration',column])
            tweets.append(tweet_limpio)
            
        
tweets_volumes, durations, tweets = zip(*sorted(zip(tweets_volumes, durations, tweets),reverse=True))
durations, tweets, tweets_volumes
dfNew2=pd.DataFrame([list(zip(tweets_volumes,durations))],columns=tweets,index=['2020-03-23'])
dfNew2
```
