```python
# -*- coding: utf-8 -*-
import pandas as pd
from googletrans import Translator
import re
from os import listdir
import shutil
import nltk
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
path_csv='./csv/'
numero_a_procesar=10
```

```python
for f in sorted(listdir(path_csv)):
    print(path_csv+f)
```

```python
def remove_single_letters(tweet):
    tweet=' '.join( [w for w in str(tweet).split() if len(w)>1] )
    return tweet
```

```python

li = []
for f in sorted(listdir(path_csv)):
    if f.endswith('.csv'):
        
        df = pd.read_csv(path_csv+f, index_col=None, header=0)
        df['date']=f[2:6]
        df['tweet'] = np.array([remove_single_letters(tweet) for tweet in df['tweet']])
        df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
        li.append(df)
        print(df)

frame = pd.concat(li, axis=0, ignore_index=True)  

frame
```

```python
to_drop = ['nan']
frame=frame[~frame['tweet'].isin(to_drop)]
#frame.dropna( inplace=True)
frame
```

```python
# Import the wordcloud library
from wordcloud import WordCloud
# Join the different processed titles together.
long_string = ','.join(list(frame['tweet'].values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
```

```python
# bigrams: get TOKENS
tokens = frame['tweet'].apply(nltk.word_tokenize) 
# Flatening nested list
flat_tokens = [term for sublist in tokens for term in sublist]

bgs = nltk.bigrams(flat_tokens)
```

```python
tokens
```

```python
# bigrams: top 10
fdist = nltk.FreqDist(bgs)
fdist_10 = fdist.most_common(15)
fdist_10
```

```python
# bigrams: source and target
labels = ['bigram', 'Weight']
df_bigrams = pd.DataFrame([tuple_item for tuple_item in fdist.items()], columns =labels)
df_bigrams[['Source_Name','Target_Name']]=pd.DataFrame([tuple_item for tuple_item in df_bigrams.bigram])
#sort in descending order
df_bigrams.sort_values(by='Weight',ascending=False).head()
```

```python
# bigrams: filter by frequency
df_bigrams_sub=df_bigrams[df_bigrams['Weight']>1] # Change your weighting filter as needed
df_bigrams_sub.head(3)
```
