```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
def isLatin(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True

```

```python
def cleanTweet(t):
    
    t=' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", t).split())
    
    e=0
    for indice, letter in enumerate(t):
        
        if letter.isupper():
            if indice!=0:
                t=t[:indice+e] + ' ' + t[indice+e:]
                e+=1
    
    
    t=split_numbers(t)
    t=clean_tweet_ngrams(t)
    return t
```

```python
def translateTweet(t):
    t=re.sub('#', '', t)
    traducido=translator.translate(t)
    #print(t, traducido.text)
    return traducido.text
```

```python
def split_numbers(t):
    t=re.sub(r"([0-9]+(\.[0-9]+)?)",r" \1 ", t).strip()
    return t
```
