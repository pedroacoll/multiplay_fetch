```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.trendinalia.com/twitter-trending-topics/spain/spain-200101.html'

r = requests.get(url)

print(r.content[:100])
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
soup
```

```python
rows = soup.select('.zebra-striped')
```

```python
filas=soup.select('.zebra-striped tbody tr')
```

```python
for fila in filas:
    name = fila.select_one('a')['title']
    duration=fila.select('div')[2].text
    print(name,duration)
```
