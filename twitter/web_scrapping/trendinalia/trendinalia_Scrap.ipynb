```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://getdaytrends.com/spain/2020-06-03/24/'

r = requests.get(url)

print(r.content[:100])
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
soup
```

```python
filas=soup.select_one('main')
print(filas)
```

```python
for fila in filas:
    name = fila.select_one('a')['title']
    duration=fila.select('div')[2].text
    print(name,duration)
```
