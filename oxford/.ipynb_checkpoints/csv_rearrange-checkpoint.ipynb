```python
import matplotlib.pyplot as plt

import numpy as np

import os

import pandas as pd
```

```python
file='./tratadas/stringency_country_date.csv'
```

```python
data = pd.read_csv(file,index_col=0)
data=data.T
data
```

```python
tuplas=[]
#print(data.columns)

for index, row in data.iterrows():
    if index!='Unnamed: 0':
        for i in range(len(data.columns)):
            print(i)
            
            if float(row[i])>0.0:
                escogido=[i]
                escogido.append(index)
                escogido.append(row[i])
                tuplas.append(escogido)
                break
    print(tuplas)       
```

```python
for index, row in data.iterrows():
    print(index)
```

```python
tuplas=sorted(tuplas)
ordenado = [tupla[1] for tupla in tuplas]
print(tuplas,ordenado)
```

```python
dataordenada=data.reindex(ordenado)

dataordenada=dataordenada.T
dataordenada
```
