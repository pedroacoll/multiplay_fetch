```python
import matplotlib.pyplot as plt

import numpy as np

import os

import pandas as pd
```

```python
file='./tratadas/stringency_country_date.csv'
```

```python
data = pd.read_csv(file,index_col=0)
data=data.T
data
```

```python
tuplas=[]
#print(data.columns)
i=0
for index, row in data.iterrows():
    
    for e in range(len(data.columns)):
        #print(data.iloc[i,e])
        if data.iloc[i,e]>0.0:
            escogido=[e]
            escogido.append(index)
            tuplas.append(escogido)
            break

    i+=1
print(tuplas)
```

```python
for index, row in data.iterrows():
    print(index)
```

```python
tuplas=sorted(tuplas)
ordenado = [tupla[1] for tupla in tuplas]
print(tuplas,ordenado)
```

```python
dataordenada=data.reindex(ordenado)

dataordenada=dataordenada.T
dataordenada
```

```python
dataordenada.to_csv('stringency_country_date_homogenea.csv')
```
